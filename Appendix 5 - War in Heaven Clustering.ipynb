```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.cluster import KMeans
```

```python
df = pd.read_excel('Features Con %.xlsx')
```

```python
df_avg = df.describe().round(2)
df_avg.to_csv("Features Con %_avg.csv")
```

```python
df_scaled = pd.DataFrame(scale(df[['Plosive%', 'Nasals%', 'Affricate%', 'Sibilant%', 'Fricative%', 'Liquid%', 'Semi Vow%']]),columns = ['Plosive%', 'Nasals%', 'Affricate%', 'Sibilant%', 'Fricative%', 'Liquid%', 'Semi Vowe%'])
df_scaled.columns
```

```python
model = KMeans(n_clusters = 7, random_state = 123)
model.fit(df_scaled)
```

```python
model.labels_
```

```python
df['Cluster'] = model.labels_.astype(int)
```

```python
df.to_csv("Features Con %_cluster.csv")
```

```python
df_clust0 = df[df['Cluster'] == 0]
df_clust0_avg = df_clust0.describe().round(2)
df_clust0_avg.to_csv("Features Con %_Clust0_avg.csv")
df_clust1 = df[df['Cluster'] == 1]
df_clust1_avg = df_clust1.describe().round(2)
df_clust1_avg.to_csv("Features Con %_Clust1_avg.csv")
df_clust2 = df[df['Cluster'] == 2]
df_clust2_avg = df_clust2.describe().round(2)
df_clust2_avg.to_csv("Features Con %_Clust2_avg.csv")
df_clust3 = df[df['Cluster'] == 3]
df_clust3_avg = df_clust3.describe().round(2)
df_clust3_avg.to_csv("Features Con %_Clust3_avg.csv")
df_clust4 = df[df['Cluster'] == 4]
df_clust4_avg = df_clust4.describe().round(2)
df_clust4_avg.to_csv("Features Con %_Clust4_avg.csv")
df_clust5 = df[df['Cluster'] == 5]
df_clust5_avg = df_clust5.describe().round(2)
df_clust5_avg.to_csv("Features Con %_Clust5_avg.csv")
df_clust6 = df[df['Cluster'] == 6]
df_clust6_avg = df_clust6.describe().round(2)
df_clust6_avg.to_csv("Features Con %_Clust6_avg.csv")
```

```python
df.head()
```

```python
# Change the clusters to labels
df['Cluster'] = df['Cluster'].replace([0], 'High Affricates')
df['Cluster'] = df['Cluster'].replace([1], 'High Plosives')
df['Cluster'] = df['Cluster'].replace([2], 'High Fricatives')
df['Cluster'] = df['Cluster'].replace([3], 'High Nasals')
df['Cluster'] = df['Cluster'].replace([4], 'High Semi Vowels')
df['Cluster'] = df['Cluster'].replace([5], 'High Sibilants')
df['Cluster'] = df['Cluster'].replace([6], 'High Liquids')
```

### Chapter 1

```python
one = df[df['Chapter'] == 1]

plt.figure(figsize=(20,50))
pal = {"High Liquids":"red", "High Affricates":"gold", "High Fricatives":"black", "High Plosives":"purple", 
       "High Sibilants":"green", "High Semi Vowels":"grey", "High Nasals":"blue"}

plot1 = sns.barplot(data = one,
             x = "Word count",
             y = "Sentence",
             hue = "Cluster",
                   palette = pal,
                   orient = 'h',
                   dodge = False)

plot1.set_title("Consonant Classifications - Chapter One: The Prelude", 
                pad=10,
                fontsize=14, 
                fontdict={"weight": "bold", "family": "sans-serif"})
plot1.set_ylabel("Sentence", fontsize=12)
plot1.set_xlabel("Word Count", fontsize=12)  
```
